```python
import pandas as pd
import numpy as np
```

```python
credits = pd.read_csv("Downloads/tmdb_5000_credits.csv")
```

```python
movies = pd.read_csv("Downloads/tmdb_5000_movies.csv")
```

```python
credits.head()
```

```python
movies.head()
```

```python
print("Credits: ", credits.shape)
print("Movies: ", movies.shape)
```

```python
credits_rename_col = credits.rename(index = str, columns = {"movie_id" : "id" })
merged_movies = movies.merge(credits_rename_col, on = 'id')
```

```python
clean_df = merged_movies.drop(columns = ['homepage', 'title_x', 'title_y', 'status', 'production_countries'])
```

```python
clean_df.head()
```

```python
print("Cleared Cols: ", clean_df.shape)
```

```python
movies_df = clean_df
```

```python
movies_df.info()
```

```python
#Weighted Average = (Rv + Cm) / v + m
# where 
# R = average rating for the movie
# v = number of votes for movie
# C = minimum votes for entry to be considered 
# m = Average rating across the whole dataset
```

```python
R = movies_df['vote_average']
v = movies_df['vote_count']
m = movies_df['vote_count'].quantile(0.69)
C = movies_df['vote_average'].mean()
```

```python
movies_df['weighted_avg'] = ((R*v) + (C*m)) / (v + m)
```

```python
movies_df.head()
```

```python
sorted_movies = movies_df.sort_values('weighted_avg', ascending = False)
```

```python
sorted_movies[['original_title', 'vote_count', 'vote_average', 'weighted_avg', 'popularity']].head(20)
```

```python
#Content-based filtering - determined by user behavior
```

```python
popularity = movies_df.sort_values('popularity', ascending = False)
popularity.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaling = MinMaxScaler()
scaled_movies = scaling.fit_transform(movies_df[['weighted_avg', 'popularity']])
normalized_movies = pd.DataFrame(scaled_movies, columns = ['weighted_avg', 'popularity'])
normalized_movies.head()                                            
```

```python
sorted_movies[['normalized_wavg', 'normalized_popularity']] = normalized_movies 
```

```python
sorted_movies.head()
```

```python
sorted_movies['score'] = sorted_movies['normalized_wavg'] * 0.5 + sorted_movies['normalized_popularity'] * 0.5
```

```python
sorted_movies.head()
```

```python
scored_movies = sorted_movies.sort_values(['score'], ascending = False)
scored_movies[['original_title', 'normalized_wavg', 'normalized_popularity', 'score']].head()
```
